```python
def search_artist_get_tracks(scraper, artist_name):
    artist_uri = scraper.get_artist_uri(artist_name)
    album_uris = scraper.get_artist_albums(artist_uri)
    tracks = []
    for album in album_uris:
        album_tracks = scraper.get_album_tracks(album)
        tracks.extend(album_tracks)
    return tracks
    
```
